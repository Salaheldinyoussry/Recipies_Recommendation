```python
import pandas as pd

df = pd.read_csv("F:\\food dataset\\RAW_interactions.csv")
```

```python
df=df.drop(['date','review'], axis = 1)
df=df.truncate(before=1, after=40000)

```

```python
df=pd.pivot_table(df,index=['user_id'],columns='recipe_id',values='rating')
df
```

```python
df=df.dropna(thresh=50,axis=1)
df
```

```python
df=df.fillna(0)
df.head()
```

```python
item_simlarity=df.corr(method='pearson')
item_simlarity
```

```python
def get_similar_rec(name,rating):
    sm=item_simlarity[name]*(rating-2.5)
#     sm=sm.rename({name:getname(name)})
#     sm=sm.rename(index={name:getname(name)})

    sm=sm.sort_values(ascending=False)
    return sm
```
